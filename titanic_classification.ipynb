```python
titanic_data.drop('Cabin', axis=1, inplace=True)
```

```python
titanic_data.dropna(inplace=True)
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data.Sex = titanic_data.Sex.map({'male':0, 'female':1})
```

```python
pd.get_dummies(titanic_data['Embarked']).head()
```

```python
embarked = pd.get_dummies(titanic_data['Embarked'], prefix = 'embarked',drop_first=True)
embarked.head()
```

```python
titanic_data = pd.concat([titanic_data,embarked],axis=1)
```

```python
titanic_data.head()
```

```python
titanic_data.drop(['PassengerId','Name','Sex','Ticket','Fare','Embarked'],axis=1,inplace=True)
```

```python
titanic_data.head()
```

```python
titanic_data.columns
```

```python
y=titanic_data['Survived']
x=titanic_data.drop(['Survived'],axis=1)
```

```python
x.columns
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 4)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_test.columns
```

```python
prediction = model.predict(x_test)
```

```python
prediction
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,prediction)
```

```python
from sklearn.metrics import classification_report
classification_report(y_test,prediction)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,prediction)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
